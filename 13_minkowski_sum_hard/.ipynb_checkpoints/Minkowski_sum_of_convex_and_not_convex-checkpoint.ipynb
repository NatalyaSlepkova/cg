# Сумма Минковского выпуклого и невыпуклого многоугольников
============================================================================================================

&#8195; В прошлом разделе (билет №12) было описано как искать сумму Минковского "одинаковых" многоугольников (все выпуклые или все невыпуклые). Здесь будет рассмотрено, как находить сумму Минковского двух различных многоугольников.

## Содержание
* Напоминание
* Общий алгоритм
* Понятие псевдодиска
* Понятие более крайнего многоугольника в заданном направлении
* Теоремы и утверждения, необходимые для доказательства теоремы о сложности алгоритма
* Теорема о сложности алгоритма
* Худший случай

## Напоминание
***

Суммой Минковского двух множеств $S_1\subset\mathbb{R^2}$ и $S_2\subset\mathbb{R^2}$ называется множество $S_1 \oplus S_2$, определенное как <br>
<div style="padding-left:250px">
$S_1 \oplus S_2 := p + q : p \in S_1, q \in S_2$ 
</div> <br>
где $p + q$ &ndash; векторная сумма двух векторов $p$ и $q$, т.е. если $p = (p_x, p_y)$ и $q = (q_x, q_y)$, то
<div style="padding-left:275px">
$p + q := (p_x + q_x, p_y + q_y)$.
</div>
***

&#8195; Пусть имеется невыпуклый многоугольник $P$ и выпуклый многоугольник $R$ с количеством вершин $n$ и $m$ соответственно.

## Общий алгоритм
***

>* Триангулируем многоугольник $P$ &ndash; как будет показано ниже, получим $n-2$ треугольника $t_1$, ..., $t_{n-2}$; <br>
>* Посчитаем сумму Минковского $R$ и каждого из полученных треугольников $t_i$ (сумма Минковского двух выпуклых многоугольников, см. билет №12). Пусть $R \oplus t_i = T_i$; <br>
>* Объединение всех $T_i$ и будет ответом. <br>

## Псевдодиски
***
&#8195; Оказывается границы двух сумм Минковского могут пересекаться только весьма специфическим образом. Чтобы уточнить как именно введем понятие $\underline{псевдодисков}$.
### Свойство псевдодисковости
<img src="images/pseudodiscs.jpg\" style=" width: 40%; height: 40%\"> <br>
Здесь *слева показаны псевдодиски, справа &ndash; нет.* <br>

&#8195; Рассмотрим два плоских объекта $o_1$ и $o_2$, каждый из которых ограничен простой замкнутой кривой. На интуитивном уровне пара $(o_1, o_2)$ называется *парой псевдодисков*, если их границы &lambda;$o_1$ и &lambda;$o_2$ пересекаются не более чем в двух точках (как на рисунке). Но в вырожденных случаях &ndash; например, когда границы пересекаются на одномерном участке, &ndash; это определение оказывается недостаточно общим. Поэтому формально будем говорить, что $(o_1, o_2)$ &ndash; пара псевдодисков, если оба множества &lambda;$o_1 \cap int(o_2)$ и &lambda;$o_2 \cap int(o_1)$ связаны (здесь $int(o)$ &ndash; внутренность объекта $o$). Набор объектов, ограниченных простыми замкнутыми кривыми, называется *набором псевдодисков*, если каждая пара объектов в нем является парой всевдодисков. Это, например, наборы кругов или осепараллельных квадратов. Отметим, что свойство псевдодисковости относится к взаимному расположению границ двух объектов &ndash; фраза "данный объект является псевдодиском" лишена смысла. <br>
&#8195; Пусть имеются два многоугольника $P$ и $P'$. Будем говорить, что точка пересечения $p \in$ &lambda;$P \cap$ &lambda;$P'$ является $\underline{перекрестьем}$ $\underline{границ}$, если в точке $p$ граница &lambda;$P$ переходит из внутренности $P'$ во внешность $P'$. <br>
### Замечание
> Пара многоугольных псевдодисков $P$ и $P'$ определяет не более дыух перекрестий границ.

## Более крайний многоугольник
***
&#8195; Пусть имеется пара выпуклых многоугольников с непересекающимися внутренностями. Будем говорить, что один многоугольник является более крайним в направлении $\vec{d}$, чем другой, если его крайние точки находятся дальше в этом направлении. Например, многоугольник будет более крайним в положительном направлении оси $x$, если его самые правые точки лежат правее самых правых точек другого многоугольника. <br>
&#8195; Будем рассматривать крайние точки в разных направлениях. Для этого представим множество всех направлений единичной окружностью с центром в начале координат: точка $p$ на единичной окружности представляет направление вектора из начала координат в $p$. Диапозон направлений между $\vec{d}_1$ и $\vec{d}_2$ определяется как множество направлений, соответствующих дуге окружности между точками, представляющими $\vec{d}_1$ и $\vec{d}_2$ в направлении против часовой стрелки. Стоит отметить, что диапозон направлений между $\vec{d}_1$ и $\vec{d}_2$ &ndash; не то же самое, что и между $\vec{d}_2$ и $\vec{d}_1$.
### Замечание 1 
> Можно заметить, что если $P$ и $R$ &ndash; два выпуклых многоугольника и $T = P \oplus R$. То крайняя точка $T$ в направлении $\vec{d}$ является суммой крайних точек $P$ и $R$ в направлении $\vec{d}$. 

#### Более наглядно
<img src="images/extremePoint.png\" style=" width: 30%; height: 30%\"> <br>
### Замечание 2
> Пусть $P_1$ и $P_2$ &ndash; выпуклые многоугольники с непересекающимися внутренностями. Пусть $\vec{d}_1$ и $\vec{d}_2$ &ndash; направления, в которых $P_1$ является более крайним, чем $P_2$. Тогда либо $P_1$ более крайний, чем $P_2$ в всех направлениях из диапозона между $\vec{d}_1$ и $\vec{d}_2$, либо более крайний во всех направлениях из диапозона между $\vec{d}_2$ и $\vec{d}_1$. 

#### Более наглядно
&#8195; *Один выпуклый многоугольник является более крайним, чем другой, в связном диапозоне направлений*
<img src="images/extremePolygon.jpg\" style=" width: 40%; height: 40%\"> <br>
>Здесь:
* $[\vec{d}_1 .. \vec{d}_2]$ $P_1$ и $P_2$ одинаково крайние;
* $(\vec{d}_2 .. \vec{d}_3)$ $P_2$ более крайний;
* $(\vec{d}_3 .. \vec{d}_1)$ $P_1$ более крайний.

## Теорема 1
***
> Пусть $P_1$ и $P_2$ &ndash; выпуклые многоугольники с непересекающимися внутренностями и $R$ &ndash; еще один выпуклый многоугольник. Тогда две *суммы Минковского* $P_1 \oplus R$ и $P_2 \oplus R$ &ndash; пара псевдодисков.

$\triangleleft$ Пусть $T_1 = P_1 \oplus R$ и $T_2 = P_2 \oplus R$. В силу симметрии достаточно доказать, что &lambda;$T_1 \cap int(T_2)$ &ndash; связное множество. 
<img src="images/theorem1.png\" style=" width: 25%; height: 25%\"> <br>
&#8195; Предположим, что &lambda;$T_1 \cap int(T_2)$ несвязно. Тогда существуют четыре направления $\vec{d}_p$, $\vec{d}_q$, $\vec{d}_r$, $\vec{d}_s$ &ndash; это направления внешних нормалей в точках $p$, $q$, $r$, $s \in$ &lambda;$T_1$, находящихся на границе &lambda;$T_1$ в указанном порядке, причем $p$, $r \in int(T_2)$, а $q$, $s \notin int(T_2)$, &ndash; такие что $T_2$ является более крайним в направлениях $\vec{d}_p$ и $\vec{d}_r$, но не в направлениях $\vec{d}_q$ и $\vec{d}_s$. Из $\underline{замечания 1}$ *о крайней точке в сумме Минковского* следует, что $P_1$ более крайний, чем $P_2$ в направлениях $\vec{d}_p$ и $\vec{d}_r$, но не в направлениях $\vec{d}_q$ и $\vec{d}_s$. Однако, это противоречит *$\underline{замечанию 2}$ о более крайнем многоугольнике в связном диапозоне направлений*. $\triangleright$ <br>

&#8195; Этот результат полезен в сочетании со следующей теоремой.

## Теорема 2
***
> Пусть $S$ &ndash; набор выпуклых многоугольных псевдодисков, имеющих в совокупности $n$ сторон. Тогда сложность их объединения не превосходит $2n$.

$\triangleleft$ Сопоставим каждую вершину объединения с вершиной псевдодиска таким образом, что с каждой вершиной псевдодиска будет сопоставлено не более двух вершин. Тем самым будет доказано, что сложность объединения не превышает $2n$. <br>
&#8195; Сопоставление производится следующим образом. На границе объединения встречаются вершины двух типов: вершины псевдодиска и точки пересечения внутренностей двух ребер псевдодиска. <br>
&#8195; Вершины первого типа просто сопоставим самим себе.
<img src="images/theorem2.png\" style=" width: 25%; height: 25%\"> 
&#8195; Теперь рассмотрим *вершину объединения* $v$, являющуюся пересечением ребра $e$ псевдодиска $P \in S$ и ребра $e'$ псевдодиска $P' \in S$. Тогда $e \cap e'$ &ndash; *перекрестье границ*. Следовательно, согласно $\underline{замечанию}$ *о перекрестьи границ многоугольных псевдодисков*, либо $e$ не имеет другого *перекрестья границ* с &lambda;$P'$, либо $e'$ не имеет другого *перекрестья границ* с &lambda;$P$ (возможно и то и другое одновременно). Без потери общности можно предположить, что $e$ не имеет другого *перекрестья границ* с &lambda;$P'$. Начав с точки $e \cap e'$, последуем вдоль $e$ во внутренность $P'$. Поскольку $e$ не пересекает &lambda;$P'$ во второй раз, мы должны достичь конца $e$ раньше, чем достигнем внешности $P'$. Этот конец $e$ мы и сопоставим вершине $v$. <br>
&#8195; Утверждается, что при таком сопопставлении каждой вершине $v$ любого псевдодиска $P$ сопоставлено не более двух *вершин объединения*. <br>
&#8195; Сначала рассмотрим случай, когда $v$ не лежит ни внутри ни на границе никакого другого псевдодиска $P'$. Тогда очевидно, что $v$ &ndash; *вершина объединения*, сопоставленная только самой себе. Пусть теперь $v$ лежит внутри какого-нибудь псевдодиска $P'$. Это означает, что $v$ лежит внутри объединения. Проследуем по обоим ребрам $P$, инцидентным $v$, пока не дойдем до границы объединения в точке *перекрестья* с каким-нибудь другим ребром; только эти два *перекрестья* (если они существуют) и сопоставлены $v$. Наконец, если $v$ лежит на границе какого-то другого псевдодиска $P'$, то $v$ может быть сопоставлена она сама и (как в случае, когда $v$ находится внутри псевдодиска) первые два *перекрестья*, встретившиеся на инцидентных ей ребрах. Но $v$ могут быть сопоставлены концы обоих инцидентных ей ребер, только если инцидентные ребра идут из $v$ во внутренность объединения псевдодисков. В таком случае $v$ не является *вершиной объединения* и не может быть сопоставлена сама себе. Следовательно, во всех случаях с $v$ сопоставлено не более двух вершин. $\triangleright$ 

## Утверждение
***
>Многоугольник $P$ можно разбить на $n - 2$ треугольника $t_1, t_2, ..., t_{n-2}$.

$\triangleleft$ Докажем данное утверждение индукцией по $n$. 
###### База
&#8195; При $n = 3$ многоугольник уже является треугольником, так что утверждение верное. <br>

###### Переход
&#8195; Предположим, что для всех $m \lt n$ $(n \gt 3)$ утверждение доказано. Докажем его для $n$. <br>

&#8195; Для начала докажем, что в многоугольнике $P$ существует диагональ. Пусть $v$ &ndash; самая левая вершина в $P$ (если таких несколько выберем самую нижнюю). Вершины $u$ и $w$ &ndash; две соседние по ребрам вершины с $v$. Тогда если отрезок $\overline{uw}$ лежит внутри $P$, то диагональ найдена.  

<img src="images/uw.png\" style=" width: 20%; height: 20%\">

&#8195; Иначе найдется хотя бы одна вершина многоугольника, лежащая внутри $\triangle{uvw}$ или на диагонали $\overline{uw}$. Пусть $v'$ &ndash; одна из таких точек, которая наиболее удалена от диагонали $\overline{uw}$. Отрезок $\overline{vv'}$ не может пересекать никакое ребро $P$. Иначе эта точка пересечения будет лежать внутри $\triangle{uvw}$ и будет самой удаленной от $\overline{uw}$, что противоречит определению $v'$. Следовательно $\overline{vv'}$ и есть диагональ.<br>

<img src="images/vv'.png\" style=" width: 25%; height: 25%\">

&#8195; Таким образом диагональ существует. Любая диагональ разбивает многоугольник $P$ на два "подмногоугольника" $P_1$ и $P_2$. Пусть $m_1$ и $m_2$ - количество вершин в $P_1$ и $P_2$ соответственно $(m_1, m_2 \lt n)$. Тогда для $P_1$ и  $P_2$ по предположению индукции утверждение выполнено. Следовательно, и $P$ тоже допускает триангуляцию. <br>
&#8195; Остается доказать, что любая триангуляция $P$ состоит из $n - 2$ треугольников. Заметим, что каждая из $(n - 2)$ вершин из $P$ (кроме двух, которые лежат на выбранной диагонали) принадлежит только одному "подмногоугольнику" ($P_1$ или $P_2$). Поэтому выполняется $m_1 + m_2 = n + 2$. Также по предположению индукции, $P_1$ содержит $(m_1 - 2)$ треугольника, а $P_2$ - $(m_2 - 2)$. Следовательно, треугольников всего $(m_1 - 2) + (m_2 - 2) = n - 2$. $\triangleright$

### Замечание 

> Какими бы ни были множества $S_1$, $S_2$ и $S_3$ для них выполнено равенство:
<div style="padding-left:120px">
$S_1 \oplus (S_2 \cup S_3) = (S_1 \oplus S_2) \cup (S_1 \oplus S_3)$
</div>

## Теорема
***
>Пусть $P$ — невыпуклый многоугольник с $n$ вершинами, а $R$ — выпуклый с $m$. Тогда сложность $P \oplus R$ равна $O(nm)$.

$\triangleleft$ Так как многоугольник $P$ невыпуклый, то его можно триангулировать на $n - 2$ треугольников $t_1$, ..., $t_{n-2}$. Применяя $\underline{замечание}$, получаем: <br>
<div style="padding-left:380px">
$P \oplus Q = \bigcup_{i=1}^{n-2}{P_i \oplus Q}$. <br>
</div>

&#8195; Поскольку $t_i$ &ndash; треугольник, а $R$ &ndash; выпуклый многоугольник с $n$ вершинами, то $t_i \oplus R$ &ndash; выпуклый многоугольник, имеющий не более $m + 3$ вершин. При этом внутренности треугольников не пересекаются, поэтому по $\underline{теореме 1}$ набор *сумм Минковского* представляет собой набор *псевдодисков*. Следовательно, по $\underline{теореме 2}$ сложность их объединения линейно зависит от суммы сложностей. Поэтому сложность $P \oplus R$ равна $O(nm)$. $\triangleright$ 
     

## Худший случай
***
&#8195; Эта верхняя оценка сложности суммы выпуклого и невыпуклого многоугольников является точной в худшем случае. Чтобы убедиться в этом, рассмотрим многоугольник $P$ с $[n / 2]$ направленными вверх пиками и гораздо меньший многоугольник $R$, являющийся верхней половиной правильного $(2m - 2)$-угольника. Их *сумма Минковского* тоже будет иметь $[n / 2]$ пиков, причем на вершине каждого из них будет находиться m вершин.
