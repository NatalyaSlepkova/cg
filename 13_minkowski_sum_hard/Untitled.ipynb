# Сумма Минковского выпуклого и невыпуклого многоугольников
=======================================================================================================

В прошлом разделе (билет №12) было описано как искать сумму Минковского <<одинаковых>> многоугольников (все выпуклые или все невыпуклые). Здесь будет рассмотрено, как находить сумму Минковского двух различных многоугольников.

Пусть имеется невыпуклый многоугольник $P$ и выпуклый многоугольник $R$ с количеством вершин $n$ и $m$ соответственно.
## Утверждение
---------------------
Многоугольик $P$ можно разбить на $n - 2$ треугольника $t_1, t_2, ..., t_{n-2}$.<br>

$\triangleright$<br><div style="padding-left:40px">
Докажем данное утверждение индукцией по $n$. 
###### База
При $n = 3$ утверждение очевидно.
###### Переход
Предположим, что для всех $m \lt n$ $(n \gt 3)$ утверждение доказано. Докажем его для $n$. 

Для начала докажем, что в многоугольнике $P$ существует диагональ. Пусть $v$ - самая левая вершина в $P$ (если таких несколько выберем самую нижнюю). Вершины $u$ и $w$ - две соседние по ребрам вершины с $v$. Тогда если отрезок $\overline{uv}$ лежит внутри $P$, то диагональ найдена. Иначе найдется хотя бы одна точка, лежащая внутри $\triangle{uvw}$ или на диагонали $\overline{uv}$. Пусть $v'$ - одна из таких точек, которая наиболее удалена от диагонали $\overline{uv}$. Отрезок $\overline{vv'}$ не может пересекать ребро $P$. Иначе эта точка пересечения будет лежать внутри $\triangle{uvw}$ и будет самой удаленной от $\overline{uv}$, что противоречит определению $v'$. Тогда $\overline{vv'}$ и есть диагональ. 

Таким образом диагональ существует. Любая диагональ разбивает многоугольник $P$ на два "подмногоугольника" $P_1$ и $P_2$. Пусть $m_1$ и $m_2$ - количество вершин в $P_1$ и $P_2$ соответственно $(m_1, m_2 \lt n)$. Тогда для $P_1$ и  $P_2$ по предположению индукции выполнено. Следовательно, и для $P$ утверждение выполняется. <br>

Остается доказать, что любая триангуляция $P$ состоит из $n - 2$ треугольников.
