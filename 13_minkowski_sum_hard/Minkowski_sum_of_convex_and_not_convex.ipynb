# Сумма Минковского выпуклого и невыпуклого многоугольников
=======================================================================================================

В прошлом разделе (билет №12) было описано как искать сумму Минковского <<одинаковых>> многоугольников (все выпуклые или все невыпуклые). Здесь будет рассмотрено, как находить сумму Минковского двух различных многоугольников.

Пусть имеется невыпуклый многоугольник $P$ и выпуклый многоугольник $R$ с количеством вершин $n$ и $m$ соответственно.

Для начала докажем следующее.
## Утверждение
---------------------
Многоугольик $P$ можно разбить на $n - 2$ треугольника $t_1, t_2, ..., t_{n-2}$.<br>

$\triangleright$<br><div style="padding-left:40px">
Докажем данное утверждение индукцией по $n$. 
###### База
При $n = 3$ многоугольник уже является треугольником, так что утверждение верное. <br>

###### Переход
Предположим, что для всех $m \lt n$ $(n \gt 3)$ утверждение доказано. Докажем его для $n$. <br>

Для начала докажем, что в многоугольнике $P$ существует диагональ. Пусть $v$ - самая левая вершина в $P$ (если таких несколько выберем самую нижнюю). Вершины $u$ и $w$ - две соседние по ребрам вершины с $v$. Тогда если отрезок $\overline{uw}$ лежит внутри $P$, то диагональ найдена.  

<img src="images/uw.png\" style=" width: 20%; height: 20%\">

<div style="padding-left:40px">
Иначе найдется хотя бы одна вершина многоугольника, лежащая внутри $\triangle{uvw}$ или на диагонали $\overline{uw}$. Пусть $v'$ - одна из таких точек, которая наиболее удалена от диагонали $\overline{uw}$. Отрезок $\overline{vv'}$ не может пересекать никакое ребро $P$. Иначе эта точка пересечения будет лежать внутри $\triangle{uvw}$ и будет самой удаленной от $\overline{uw}$, что противоречит определению $v'$. Следовательно $\overline{vv'}$ и есть диагональ.

<img src="images/vv'.png\" style=" width: 20%; height: 20%\">

<div style="padding-left:40px">
Таким образом диагональ существует. Любая диагональ разбивает многоугольник $P$ на два "подмногоугольника" $P_1$ и $P_2$. Пусть $m_1$ и $m_2$ - количество вершин в $P_1$ и $P_2$ соответственно $(m_1, m_2 \lt n)$. Тогда для $P_1$ и  $P_2$ по предположению индукции утверждение выполнено. Следовательно, и $P$ тоже допускает триангуляцию. <br>

Остается доказать, что любая триангуляция $P$ состоит из $n - 2$ треугольников. Заметим, что каждая из $(n - 2)$ вершин из $P$ (кроме двух, которые лежат на выбранной диагонали) принадлежит только одному "подмногоугольнику" ($P_1$ или $P_2$). Поэтому выполняется $m_1 + m_2 = n + 2$. Также по предположению индукции, $P_1$ содержит $(m_1 - 2)$ треугольника, а $P_2$ - $(m_2 - 2)$. Следовательно, треугольников всего $(m_1 - 2) + (m_2 - 2) = n - 2$. <br>
</div>$\triangleleft$

Заметим следующий факт: какими бы ни были множества $S_1$, $S_2$ и $S_3$ для них выполнено равенство:
<div style="padding-left:140px">
$S_1 \oplus (S_2 \cup S_3) = (S_1 \oplus S_2) \cup (S_1 \oplus S_3)$
</div>

## Теорема
---------------
>Пусть $P$ — невыпуклый многоугольник с $n$ вершинами, а $Q$ — выпуклый с $m$. Тогда количество вершин в $P \oplus Q$ есть $O(nm)$.

<div style="padding-left:40px">
$\triangleleft$ <br>
Возьмем триангуляцию $P$, для каждого $P_i$ посчитаем его сумму и объединим. <br>
Более формально: $P \oplus Q = \bigcup_{i=1}^{n-2}{P_i \oplus Q}$. <br>

  Исходя из оценки сложности суммы двух выпуклых полигонов, каждое объединение $P_i \oplus Q$ будет иметь максимум $m + 3$ ребер. Поскольку треугольники триангуляции не пересекаются, то ${P_i \oplus Q}$ есть набор псевдодисков. Объединение псевдодисков линейно по вершинам/ребрам, значит общая сложность $P \oplus Q$ есть $O(nm)$. $\triangleright$ </div>
